```python
import pandas as pd
import datetime

date = datetime.datetime.now().strftime("%Y-%m-%d")
today = pd.read_csv(f'dataSource/unhandled/A-{date}.csv')
increased_year = today[today["年初至今涨跌幅"] > 0]

print(f"年线为正 {len(increased_year)}")
print(today.info())
```

```python
increased_year_simple = increased_year[['代码', '名称','涨跌幅', '最新价', '涨跌额', '市盈率-动态', '总市值', '流通市值', '年初至今涨跌幅']].copy()

# increased_year_today = increased_year_simple[increased_year['涨跌幅'] > 1]




increased_year_today = increased_year_simple[\
                            (increased_year['涨跌幅'] > 1) &\
                            (increased_year['市盈率-动态'] > 0) &\
                            (increased_year['总市值'] / increased_year['流通市值'] < 2)]

increased_year_today = increased_year_today.sort_values(by = "涨跌幅", ascending = False)

increased_year_today.to_csv(f'increased_year_today/{date}.csv')

increased_year_today
```
